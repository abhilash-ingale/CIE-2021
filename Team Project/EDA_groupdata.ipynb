```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# User defined functions 

# 1. Plot Data
def plot_data(paths,title):
    # Load dataframes
    df_temp = pd.read_csv(paths)

    t = df_temp["Time (s)"].to_numpy()
    gyro_x = df_temp["Gyroscope x (rad/s)"].to_numpy()
    gyro_y = df_temp["Gyroscope y (rad/s)"].to_numpy()
    gyro_z = df_temp["Gyroscope z (rad/s)"].to_numpy()

    # plot Gyroscope data
    fig, axs = plt.subplots(3)
    fig.suptitle(title)
    axs[0].plot(t, gyro_x)
    axs[1].plot(t, gyro_y)
    axs[2].plot(t, gyro_z)
```
