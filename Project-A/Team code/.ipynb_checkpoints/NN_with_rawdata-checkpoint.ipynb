```python
########################### User-defined fucntions###############################

# 1. read_from_directory()

# This function sorts the list of directories into lists of up and down directories
# Input: filepath (string)
# Output: train_subjects, test_subjects

def read_from_directory(path,ratio): 
    #path = 'C:/Users/admin/Documents/01_RWTH Courses/03_ WiSe21/CIE/Project A/Data/All data/Smartphone1Copy/'

    if (os.path.isdir(path) and (not os.path.exists(path))):
            print("ERROR: Directory does not exist")

    else:
        # directories variable contains all folders under the path directory
        directories = os.listdir(path)

        # Return error message when no sub-directories are present
        if len(directories) == 0:
            print("ERROR: No sub-directories found in the given directory")
                # code to exit loop somehow! 

        subjects = []    

        # Check keywords and segregate accordingly
        for entry in directories: 
            split_list = entry.split('_')
            for i in split_list:
                if ('subject' in i):
                    subjects.append(i)   

        def check_last_three(i):
            return i[-3:]

        subjects = sorted(list(set(subjects)),key=check_last_three)
        #print(subjects)
        
        # 
        train_subjects = random.sample(subjects, m.floor(ratio*len(subjects)))
        #print(len(train_subjects))
    
        test_subjects = list(set(subjects)-set(train_subjects))
        #print(len(test_subjects))
        
    return train_subjects, test_subjects, directories


# 2. prepare_df()

def prepare_df(train_ls,test_ls,directories):
    keys = ['upstairs','downstairs','normal']
    cols = ['Gyroscope x (rad/s)','Gyroscope y (rad/s)','Gyroscope z (rad/s)','Label']
    
    train_df = pd.DataFrame(columns = cols)
    test_df = pd.DataFrame(columns = cols)
    
    for i in train_ls:
        for key in keys: 
            name = i +'_'+ key +'/Gyroscope.csv'
            
            if (name in directories):
                df = pd.read_csv(path + name).drop('Time (s)', axis=1)
                df['Label'] = keys.index(key)
                train_df.appned(df) 
                print(train_df)
    
    for j in test_ls:
        for key in keys: 
            name = j + '_'+ key +'/Gyroscope.csv'
            
            if (name in directories):
                df = pd.read_csv(path+name).drop('Time (s)', axis=1)
                df['Label'] = keys.index(key)
                train_df.appned(df) 
    
    return train_df, test_df

```
